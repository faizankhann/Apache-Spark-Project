```python
import pyspark
from pyspark import SQLContext

sc = pyspark.SparkContext()
sql = SQLContext(sc)
```

```python
# reading csv file
df = (sql.read
         .format("com.databricks.spark.csv")
         .option("header", "true")
         .load("time_series_19-covid-Confirmed_archived_0325.csv"))

```

```python
# getting rid of extra columns
columns_to_drop = ['Province/State','Lat','Long']
df = df.drop(*columns_to_drop)

# selecting only Austrailia Region
df2 = df.where(df['Country/Region'] == 'Australia')

# summing up cases of selected region
exprs = {x: "sum" for x in df.columns if x is not df.columns[0]}
df2=df2.groupBy("Country/Region").agg(exprs)

# renaming dates to integers 0, 1, 2 ....
newNames = []
for count,ele in enumerate(df2.columns,-1):
    if(count==-1):
        # newNames.append('Country/Region')
        newNames.append('Index')
        continue
    newNames.append(count)
    # newNames.append(str(count))

for c, n in zip(df2.columns[0:], newNames):
    df2=df2.withColumnRenamed(c, str(n))

# converting to pandas dataframe to take transpose
df2=df2.toPandas().set_index("Index").transpose() 

df2['Dates'] = df2.index # New column for Dates
df2=df2.rename(columns={"Australia": "Cases"})

# print((df2["Cases"]))
# print((df2["Dates"]))

# from pyspark.sql.types import *

# # making schema for converting back to pyspark dataframe
# mySchema = StructType([ StructField("Cases", DoubleType(), True)\
#                         ,StructField("Dates", StringType(), True)\
#                        ])

# # converting back to pyspark dataframe
# df2 = sql.createDataFrame(df2,mySchema)
# # print(type(df2))
# # df2.show()
```

```python
df_test = df2[9:60]
import pandas as pd
# print(df2)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
    print(df_test)
```

```python
df_test = df2[:60] # excluding last two dates because we have to predict value for last to dates
print(df_test)
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model

#Define the problem
# problem = [75, 104, 147, 204, 275]
problem = df_test['Cases']

#create x and y for the problem

x=[]
y=[]

for (xi, yi) in enumerate(problem):
  features=[]
  for i in range(3):
    features.append(pow(xi,i))
  x.append(features)
  y.append(yi)
# print(x)
# print(y)
x=np.array(x)
y=np.array(y)

# Create linear regression object
regr = linear_model.LinearRegression()
regr.fit(x, y)

#create the testing set
x_test_range=range(len(x),3+len(x))
x_test=[]
for xi in x_test_range:
  features=[]
  for i in range(3):
    features.append(pow(xi,i))
  x_test.append(features)
# print(x_test)

# The coefficients
print('Coefficients: \n', regr.coef_)
print('Intercept: \n', regr.intercept_)
# The mean squared error
print("Mean squared error: %.2f"
              % np.mean((regr.predict(x) - y) ** 2))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x,y))

#Do predictions
y_predicted = regr.predict(x_test)

print("Next few numbers in the series are")
for pred in y_predicted:
  print(pred)

plt.scatter(range(len(problem)), problem, color='black')
plt.scatter(x_test_range, y_predicted,  color='red')

plt.show()
```
